## คำถาม

1. เมื่อกำหนดให้สมาชิกในคลาสให้มี Access Modifier ดังต่อไปนี้ จะสามารถเข้าถึงจากที่ไหนได้บ้าง


1.1 public

เข้าถึงได้จากทุกที่ ไม่มีข้อจำกัดใดๆ

1.2 private

เข้าถึงได้เฉพาะภายในคลาสเดียวกันเท่านั้น

1.3 protected

เข้าถึงได้เฉพาะในคลาสเดียวกันและคลาสลูก (subclass / derived class)

1.4 internal

เข้าถึงได้เฉพาะภายใน Assembly (โปรเจกต์เดียวกัน)

1.5 protected internal

 เข้าถึงได้ใน Assembly เดียวกัน หรือจากคลาสลูกที่อยู่ต่าง Assembly
