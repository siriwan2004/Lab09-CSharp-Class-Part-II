## คำถาม

1. Member constant คืออะไร และมีคุณสมบัติอย่างไร


Member Constant คือ ค่าคงที่ที่กำหนดไว้ในคลาสหรือโครงสร้าง (struct) ซึ่งค่าไม่สามารถเปลี่ยนแปลงได้หลังจากถูกกำหนดค่าแล้ว

คุณสมบัติของ Member Constant:

เป็นค่าคงที่ (Constant) เปลี่ยนแปลงไม่ได้

ต้องกำหนดค่าเริ่มต้นตอนประกาศเสมอ

ใช้ const keyword

มีค่าเหมือนกันทุก Instance ของคลาส

2. เราสามารถประกาศ member constant ในภาษา C# ได้อย่างไร


ใช้ const นำหน้าชื่อ Field

ต้องกำหนดค่า ตอนประกาศเลย (ไม่สามารถเปลี่ยนค่าในภายหลัง)

3. member constant สามารถสร้างเป็นชนิดข้อมูลใดได้บ้าง


Member Constant สามารถใช้กับ ชนิดข้อมูลพื้นฐาน (Primitive Data Types) เช่น

จำนวนเต็ม: int, long, byte, sbyte, short, ushort, uint, ulong

จำนวนจริง: float, double, decimal

ตัวอักษรและข้อความ: char, string

ค่าเงื่อนไข: bool

Enumeration: enum

4. เราสามารถเข้าถึง member constant ได้อย่างไร


Console.WriteLine(MyClass.Pi); //เรียกใช้โดยตรงผ่านชื่อคลาส

Member Constant เป็น Static โดยปริยาย → เรียกผ่านชื่อคลาสได้เลย



5. member constant เป็น static โดยปริยายหรือไม่


ใช่ const มีพฤติกรรมเหมือน Static Field แต่ ไม่ต้องใช้ static keyword

6. มีข้อจำกัดอะไรบ้างในการใช้ member constant 


ไม่สามารถเปลี่ยนค่าได้หลังจากกำหนดแล้ว

ต้องกำหนดค่าคงที่ตอนประกาศเสมอ

ใช้ได้เฉพาะกับชนิดข้อมูลพื้นฐาน (Primitive Types)

ไม่สามารถใช้กับ DateTime, List<>, Dictionary<> หรือ Object อื่น ๆ ได้

ไม่สามารถอ้างถึงค่าของตัวแปรอื่น หรือผลลัพธ์จาก Method ได้

7. member constant เหมาะกับการใช้งานแบบไหน


ใช้กับค่าที่ไม่เปลี่ยนแปลง เช่น

ค่าทางคณิตศาสตร์ (Math.PI)

ค่าสถานะ (ERROR_CODE)

ค่าต่าง ๆ ที่ใช้ในเงื่อนไข (MAX_USER_LIMIT)

8. จงตัวอย่าง Member Constant ในไลบรารี่ของภาษา C#
